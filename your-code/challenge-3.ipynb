```python
amount_spent_customers = orders.groupby("CustomerID")["amount_spent"].agg("sum")
```

```python
vip_customers = amount_spent_customers[amount_spent_customers >= amount_spent_customers.quantile(0.95)]
```

```python
orders["VIP or Preferred"] = ["VIP" if order in vip_customers else "Preferred" for order in orders["CustomerID"]]
```
